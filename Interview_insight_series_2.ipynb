# Interview insight series 2

# ** # Q1. Understanding User Needs Through Research**

You’re managing a student productivity app. After a recent update, daily active users have dropped by 30%.
How would you use user research to identify and solve the problem? Include how and when you'd apply different types of research methods, and what frameworks or tools you'd use.

*   Ans) Problem Statement: Daily Active Users (DAU) dropped by 30% following a recent update to a student productivity app.*

# Phase 1: Quantitative Analysis (The "Where")
Before talking to users, we must identify exactly where the leakage is happening in the product funnel.

Funnel Analysis: Use tools like Mixpanel or Amplitude to see where students drop off. Is it at the login screen, the new dashboard, or a specific feature like the "Study Timer"?

Cohort Analysis: Compare the behavior of "Power Users" versus "New Users" to see if the update alienated the core audience or failed to convert new ones.

Crash Reporting: Check Firebase or Sentry to ensure the drop isn't due to technical bugs or latency issues introduced in the update.

# Phase 2: Qualitative Research (The "Why")
Once the friction point is located, we use research to understand the sentiment behind the data.

User Feedback Audit: Analyze recent App Store/Play Store reviews and support tickets. Look for recurring themes like "too complex," "too many ads," or "missing my old data."

Contextual Inquiry / Student Interviews: * Method: Observe 5–10 students using the app in their natural environment (library/dorm).

Application: Students have high cognitive loads. If the update added "clicks" to reach a core feature, they will abandon it.

# Usability Testing:

Method: Use UserTesting or Lookback to give users a specific task (e.g., "Add a chemistry exam to your calendar").

Application: Identify "critical incidents" where the UI prevents the user from completing their task.

# Phase 3: Frameworks for Solutioning
To turn research into an actionable roadmap, apply these frameworks:

1) The "5 Whys": Iteratively ask "Why" to get to the root cause (e.g., Drop in DAU $\rightarrow$ Students aren't setting tasks $\rightarrow$ The 'Add Task' button is hidden $\rightarrow$ Root: The UI redesign prioritized aesthetics over utility).

2) RICE Prioritization: Once fixes are identified, rank them by Reach, Impact, Confidence, and Effort to decide what to fix first.

3) JTBD (Jobs-to-be-Done): Ensure the app is still performing the "job" it was hired for (e.g., "Help me pass my finals") rather than just looking modern.

## **Q2. Balancing Business and User Goals**

You’re the PM of an e-commerce platform. The business team wants to push flash sales aggressively, but users complain of a cluttered homepage and difficulty finding regular items. How would you balance both perspectives and move forward?  Explain how you’d apply problem-solving principles, user empathy, and data-backed insights to navigate this challenge.

* Ans) Problem Statement: Balancing the business need for "Flash Sales" with the user need for "Findability and Clarity."*

# 1. Problem-Solving Principle: Data-Backed Neutrality
Hypothesis: A cluttered homepage increases the "Time to First Action," which eventually lowers the conversion rate for both regular items and flash sales.

Action: Conduct an A/B Test.

Variant A: Aggressive flash sale banners (Current Business Request).

Variant B: A dedicated "Deals" tab or a persistent but slim "Smart Bar" for sales, keeping the search bar and categories above the fold.

# 2. User Empathy: Cognitive Load Management
Users coming to an e-commerce platform are often on a mission. If we disrupt that mission, we lose their trust.

Design Solution: Implement Intent-Based UI. If a user types in the search bar, the "Flash Sale" banners should minimize to provide a clear path to their desired product.

Segmentation: Do not push "Flash Sales" to everyone. Use data to show them only to price-sensitive users or those who have historically engaged with sales.

# 3. Business Alignment: The "North Star" Metric
The Argument: While flash sales drive short-term revenue, a poor user experience leads to Churn.

Heatmapping: Use Hotjar or Microsoft Clarity to show the business team how many users are struggling to find the "Search" button because of the banners.

Compromise: Offer high-visibility "Flash Sale" slots in the post-purchase "Thank You" page rather than hijacking the primary navigation space.
